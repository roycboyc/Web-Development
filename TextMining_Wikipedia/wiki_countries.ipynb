```python
from bs4 import BeautifulSoup
import numpy as np
import urllib.request
```

```python
with urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_sovereign_states#List_of_states') as response:
   f = response.read()
```

```python
soup = BeautifulSoup(f,'html.parser')
```

```python
print (soup) #looks like the first tr isnt what i need, but the second is
```

```python
firstArray=soup.find_all("tr") #start w 4
```

```python
firstArray[235]
```

```python
firstArray[4].get_text().replace('\xa0', ' ').replace('\nA', '').replace('\n', '').replace('None','')
```

```python
len(firstArray)
```

```python
rows=[]
```

```python
for i in range(4,236): #I counted, it turned out to be that 4-235 are relevant entries
    row=[]
    entry=firstArray[i]
    entryElements=entry.find_all("td".split())
    for element in entryElements:
        clearElement=element.get_text()
        clearElement = clearElement.replace('\xa0', ' ').replace('\nA', '').replace('\n', '').replace('A None','None')
        row.append(clearElement)
    rows.append(row)
```

```python
import csv

with open('RCohen_countries.csv', 'w') as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
    wr.writerows(rows)
```
