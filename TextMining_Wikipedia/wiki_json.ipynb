```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
with urllib.request.urlopen('https://en.wikipedia.org/wiki/Neighborhoods_in_New_York_City#Community_areas') as response:
   f = response.read()
```

```python
soup = BeautifulSoup(f,'html.parser')
```

```python
print (soup) #looks like the first tr isnt what i need, but the second is
```

```python
firstArray=soup.find_all("tr")
```

```python
rows=[]
```

```python
for i in range(1,61): #I counted, it turned out to be 61 entries, incl. summary of NYC as a whole
    row=[]
    entry=firstArray[i]
    entryElements=entry.find_all("td".split())
    for element in entryElements:
        clearElement=element.get_text()
#        if clearElement[0:3]=='Broo' or clearElement[0:3]=='Manh' or clearElement[0:3]=='Bron' or clearElement[0:3]==Stat:
        clearElement = clearElement.replace('\xa0', ' ')
        row.append(clearElement)
    rows.append(row)
```

```python
header=['Community Board (CB)', 'Area km squared', 'Pop. Census 2010', 'Pop km squared', 'Neighborhoods']
rows.insert(0,header)
```

```python
import csv

with open('RCohen_submission.csv', 'w') as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
    wr.writerows(rows)
```
